# Merge data from each CSV file into one CSV file

```python
path =r'path\Sales_Data'
list_of_files = []

for root, dirs, files in os.walk(path):
    for file in files:
        list_of_files.append(os.path.join(root,file))
for name in list_of_files:
    print(name)
```

```python
df_all=pd.DataFrame()
for i in list_of_files:
    df=pd.read_csv(r"{}".format(i))
    df_all=pd.concat([df_all,df])
```

##### Question 5: What products are most often sold together?

```python
df_1=df_all[df_all.duplicated(subset=['Order ID'],keep=False)]
```

```python
df_1.sort_values(by=['Order ID']).head(12)
```

```python
df_1['grouped']=df_1.groupby('Order ID')['Product'].transform(lambda x:' ,'.join(x))
```

```python
df_1.head(10)
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df_1['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
